```python
import pymysql
conn = pymysql.connect(host = 'localhost',port=3306,user='root',passwd='123456',db='student',charset='utf8') ##连接信息
cursor = conn.cursor()
##a)	列查询：单独挑选出列名为“grade”的数据；
sql_a = "select grade from data1"
cursor.execute(sql_a)
result_a = cursor.fetchall()
print("第a题的答案为：{}".format(result_a))
##b)	行查询：查询name等于“吴”的所有数据；
sql_b = "select * from data1 where name = '吴'"
cursor.execute(sql_b)
result_b = cursor.fetchall()
print("第b题的答案为：{}".format(result_b))
##c)	在表中添加一行数据，数据为(999,吴,2022,19,5,84,84,88,	3,99,18)
sql_c = "insert into data1 (id,name,grade,class1,class2,class3,class4,class5,class6,class7,class8)\
        values('999','吴','2022','19','5','84','84','88','3','99','18')"
cursor.execute(sql_c)
conn.commit()   ####插入数据别忘记了进行提交操作
print('数据插入成果')
##d)	条件查询：查询name等于“吴”和列名等于“grade”的数据；
sql_d = "select grade from data1 where name = '吴'"
cursor.execute(sql_d)
result_d = cursor.fetchall()
print("第d题的答案为：{}".format(result_d))
##e)	条件查询：查询2020级中class1的成绩大于95的学生的id和name
sql_e = "select id,name from data1 where grade > 95"
cursor.execute(sql_e)
result_e = cursor.fetchall()
print("第e题的答案为：{}".format(result_e))
##f)	求平均：计算2020级class1~class8所有科目的平均值；
sql_f = "select avg(class1), avg(class2),avg(class3),avg(class4),avg(class5),avg(class6),avg(class7),avg(class8) from data1 where grade=2020"
cursor.execute(sql_f)
result_f = cursor.fetchall()
print("第f题的答案为：{}".format(result_f))
##g)	分组求平均：计算所有年级class1~class8所有科目的平均值；
sql_g = "select grade,avg(class1), avg(class2),avg(class3),avg(class4),avg(class5),avg(class6),avg(class7),avg(class8) from data1 group by grade"
cursor.execute(sql_g)
result_g = cursor.fetchall()
print("第g题的答案为：{}".format(result_g))
cursor.close()
conn.close()
```
