```python
import pymysql
```

```python
conn = pymysql.connect(host = "localhost" , user = "root" , password = "1024" , 
                       db = "作业" , charset='utf8') ##连接
```

```python
cur = conn.cursor()  # 游标
sql = "select grade from data3"
cur.execute(sql)
result = cur.fetchall()
print(result)
```

```python
cur = conn.cursor()  # 游标
sql = "select * from data3 where name='吴'"
cur.execute(sql)
result = cur.fetchall()
print(result)
```

```python
cur = conn.cursor()  # 游标
sql = '''INSERT INTO data3 (id,name,grade,class1,class2,class3,class4,class5,class6,class7,class8) VALUES (999,'吴',2022,19,5,84,84,88,3,99,18)
'''
cur.execute(sql)
conn.commit()
sql1 = "select * from data3 where id = 999"
cur.execute(sql1)
result = cur.fetchall()
print(result)
```

```python
cur = conn.cursor()  # 游标
sql = "select grade,name from data3 where name='吴'"
cur.execute(sql)
result = cur.fetchall()
print(result)
```

```python
cur = conn.cursor()  # 游标
sql = "select id,grade,name,class1 from data3 where grade=2020 and class1>95"
cur.execute(sql)
result = cur.fetchall()
print(result)
```

```python
cur = conn.cursor()  # 游标
sql = "select grade,avg(class1+class2+class3+class4+class5+class6+class7+class8) from data3  where grade=2020"
cur.execute(sql)
result = cur.fetchall()
print(result)
```

```python
cur = conn.cursor()  # 游标
sql = "select grade,avg(class1+class2+class3+class4+class5+class6+class7+class8) from data3 group by grade"
cur.execute(sql)
result = cur.fetchall()
print(result)
```

```python
cur = conn.cursor()  # 游标
sql = "select avg(class1+class2+class3+class4+class5+class6+class7+class8) from data3"
cur.execute(sql)
result = cur.fetchall()
print(result)
```
